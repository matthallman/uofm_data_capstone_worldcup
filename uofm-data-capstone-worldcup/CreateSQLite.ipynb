```python
from sqlalchemy import create_engine, inspect

import pandas as pd
import numpy as np
import glob
import os
import matplotlib.pyplot as plt
```

```python
RoundOfSixteen = "Database/Data CSV/USArd8.csv"
RoundOfEight = "Database/Data CSV/USArd16.csv"
Semis = "Database/Data CSV/USArdsemis.csv"
Finals = "Database/Data CSV/USArdFINALS.csv"
```

```python
SixteenDF = pd.read_csv(RoundOfSixteen)
EightDF = pd.read_csv(RoundOfEight)
SemisDF = pd.read_csv(Semis)
FinalsDF = pd.read_csv(Finals)
```

```python
CombinedDF = pd.concat([SixteenDF, EightDF, SemisDF, FinalsDF], axis=0)
CombinedDF.tail()
```

```python
database_path = "worldcupdb.sqlite"
conn_string = f"sqlite:///{database_path}"

# Create an engine that can talk to the database
engine = create_engine(conn_string)
```

```python
CombinedDF.to_sql('worldcupdb', con=engine)
```

```python
engine.execute("SELECT * FROM worldcupdb").fetchall()
```

```python
# get all tables
inspector_gadget = inspect(engine)
tables = inspector_gadget.get_table_names()
for table in tables:
    print(table)
    
    # get all columns in table
    columns = inspector_gadget.get_columns(table)
    for column in columns:
        print(column)
    print()
```
