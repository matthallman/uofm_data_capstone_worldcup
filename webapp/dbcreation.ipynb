```python
from sqlalchemy import create_engine, inspect

import sqlalchemy
import pandas as pd
import numpy as np
import glob
import os
import matplotlib.pyplot as plt
```

```python
allknockoutmatches = "Resources/all_knockout_matches.csv"
wincount = "Resources/wincount.csv"
odds = "Resources/odds.csv"
```

```python
allknockoutmatchesDF = pd.read_csv(allknockoutmatches,encoding='Windows-1252')
allknockoutmatchesDF = allknockoutmatchesDF.loc[allknockoutmatchesDF.Simulation<100]
```

```python
database_path = "knockoutdb2.sqlite"
conn_string = f"sqlite:///{database_path}"

# Create an engine that can talk to the database
engine = create_engine(conn_string)
```

```python
allknockoutmatchesDF.to_sql('knockoutdb', con=engine, if_exists='replace')
```

```python
engine.execute("SELECT * FROM knockoutdb").fetchall()
```

```python
# get all tables
inspector_gadget = inspect(engine)
tables = inspector_gadget.get_table_names()
for table in tables:
    print(table)
    
    # get all columns in table
    columns = inspector_gadget.get_columns(table)
    for column in columns:
        print(column)
    print()
```

```python
con = engine.connect()

query = """
    SELECT *  
    FROM knockoutdb
    WHERE "Match Winner"="United States"
    ;
    """

df = pd.read_sql(query, con)
df
```
